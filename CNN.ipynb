```
from keras.models import Sequential
from keras import layers
from keras.utils import np_utils

import numpy as np
from sklearn.utils import shuffle
import h5py
from sklearn.metrics import f1_score
```

```
def read_Data(file_path):
    data = h5py.File(file_path, 'r')
    X_train = data['trainSignal'][:]
    y_train = data['trainTarget_class'][:]
    X_test = data['testSignal'][:]
    y_test = data['testTarget_class'][:]
    data.close()

    X_train, y_train = shuffle(X_train, y_train, random_state=0)
    X_test, y_test = shuffle(X_test, y_test, random_state=0)

    return X_train, X_test, y_train, y_test
```

```
path = '/gdrive/My Drive/Intership/Task 1/Data/Datasets/sigma_0/data_20.hdf5'
X_train, X_test, y_train, y_test = read_Data(path)
```

```
# reshape data for CNN
y_train = np_utils.to_categorical(y_train, 3)
X_train = np.reshape(X_train, (1080, 128, 1))
X_test = np.reshape(X_test, (120, 128, 1))

print('\nX_train shape: ', X_train.shape,
      '\ny_train shape: ', y_train.shape,
      '\nX_test shape: ', X_test.shape,
      '\ny_test shape: ', y_test.shape
      )
```

```
verbose, epochs, batch_size = 2, 200, 100
n_features, n_chanels, n_outputs = X_train.shape[1], X_train.shape[2], y_train.shape[1]
```

```
my_model = Sequential()
my_model.add(layers.Conv1D(filters=100,
                           kernel_size=20,
                           activation='relu',
                           input_shape=(n_features, n_chanels)))
my_model.add(layers.Dropout(0.5))
my_model.add(layers.MaxPooling1D(2))
my_model.add(layers.Conv1D(100, 5, activation='relu'))
my_model.add(layers.Flatten())
my_model.add(layers.Dense(100, activation='relu'))
my_model.add(layers.Dense(n_outputs, activation='softmax'))

my_model.summary()
my_model.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])
```

```
my_model.fit(X_train,
             y_train,
             epochs=epochs,
             batch_size=batch_size,
             verbose=verbose)
```

```
predictions = my_model.predict_classes(X_test)
print(predictions)
f1 = f1_score(y_test, predictions, average='micro')
print(f1)
```

```
def CNN(model, X_train, y_train, X_test, y_test, batch_size, epochs, verbose):
    y_train = np_utils.to_categorical(y_train, 3)
    X_train = np.reshape(X_train, (1080, 128, 1))
    X_test = np.reshape(X_test, (120, 128, 1))
    
    model.fit(X_train,
             y_train,
             epochs=epochs,
             batch_size=batch_size,
             verbose=verbose)
    predictions = model.predict_classes(X_test)
    f1 = f1_score(y_test, predictions, average='micro')
    return f1


def CNN_Quantile(model, batch_size, epochs, verbose, dir_path):
    IQR_array = []
    for i in range(5):
        f1 = []
        for j in range(30):
            path = dir_path + '%s/data_%s.hdf5' % (i, j)
            X_train, X_test, y_train, y_test = read_Data(path)
            f1.append(CNN(model, X_train, y_train, X_test, y_test, batch_size, epochs, verbose=0))
            print(str(i), '_________',str(j))
        IQR_array.append(f1)

    return IQR_array
```

```
quantiles = CNN_Quantile(my_model, batch_size, epochs, verbose, '/gdrive/My Drive/Intership/Task 1/Data/Datasets/sigma_')
np.savetxt('quantiles_MLP.txt', quantiles)
```
